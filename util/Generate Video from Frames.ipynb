```python
import glob
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ls_frames = glob.glob('out/*')
ls_frames[:5]
```

```python
ls_images = map(lambda x: cv2.imread(x), ls_frames)
plt.imshow(ls_images[0])
```

```python
len(ls_frames)
```

```python
ls_resized_images = map(lambda x: cv2.resize(x, (3840, 1920)), ls_images[:100])
```

```python
plt.imshow(ls_resized_images[0])
```

```python
OUTPUT_FILE = 'output_3.mp4'
FPS = 10
# fourcc = cv2.cv.CV_FOURCC('m', 'p', '4', 'v') # note the lower case

# http://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_video_display/py_video_display.html
fourcc = cv2.VideoWriter_fourcc(*'XVID') 
h, w, z = ls_resized_images[0].shape
```

```python
out = cv2.VideoWriter(OUTPUT_FILE,fourcc, FPS, (w,h))
```

```python
for image in ls_resized_images:
    out.write(image)

out.release()
```
